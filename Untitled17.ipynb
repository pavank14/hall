```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings("ignore")
```

```python

# Load the data

data = pd.read_csv('pune.csv')
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
data['date_time'] = pd.to_datetime(data['date_time'])
```

```python
data.set_index('date_time', inplace=True)
```

```python
data.describe()
```

```python
plt.figure(figsize=(14,7))
plt.subplot(2, 1, 1)
plt.plot(data['maxtempC'], label='Max Temperature (C)')
plt.plot(data['mintempC'], label='Min Temperature (C)')
plt.title('Temperature Over Time')
plt.legend()
```

```python
plt.subplot(2, 1, 2)
plt.plot(data['totalSnow_cm'], label='Total Snowfall (cm)')
plt.title('Snowfall Over Time')
plt.legend()
```

```python
plt.tight_layout()
plt.show()
```

```python
result = seasonal_decompose(data['maxtempC'], model='multiplicative', period=365)
result.plot()
plt.show()
```

```python
result = seasonal_decompose(data['mintempC'], model='multiplicative', period=365)
result.plot()
plt.show()
```

```python
# Train-test split
train = data[:-30]
test = data[-30:]
```

```python
model_maxtemp = ExponentialSmoothing(train['maxtempC'], seasonal='multiplicative', trend='add', seasonal_periods=365).fit()
model_mintemp = ExponentialSmoothing(train['mintempC'], seasonal='multiplicative', trend='add', seasonal_periods=365).fit()
```

```python
# Forecast for the next 5 days
forecast_maxtemp = model_maxtemp.forecast(5)
forecast_mintemp = model_mintemp.forecast(5)
```

```python
# Plot the results
plt.figure(figsize=(14,7))
plt.subplot(2, 1, 1)
plt.plot(train.index, train['maxtempC'], label='Train Max Temperature')
plt.plot(test.index, test['maxtempC'], label='Test Max Temperature')
plt.plot(forecast_maxtemp.index, forecast_maxtemp, label='Forecast Max Temperature')
plt.legend()
```

```python

# Calculate performance metrics
mae_maxtemp = mean_absolute_error(test['maxtempC'][:5], forecast_maxtemp)
mse_maxtemp = mean_squared_error(test['maxtempC'][:5], forecast_maxtemp)
mae_mintemp = mean_absolute_error(test['mintempC'][:5], forecast_mintemp)
mse_mintemp = mean_squared_error(test['mintempC'][:5], forecast_mintemp)

print(f"Max Temperature - MAE: {mae_maxtemp}, MSE: {mse_maxtemp}")
print(f"Min Temperature - MAE: {mae_mintemp}, MSE: {mse_mintemp}")

# Output the forecasts
forecast = pd.DataFrame({
    'date_time': forecast_maxtemp.index,
    'forecast_maxtempC': forecast_maxtemp.values,
    'forecast_mintempC': forecast_mintemp.values
})
print(forecast)
```
